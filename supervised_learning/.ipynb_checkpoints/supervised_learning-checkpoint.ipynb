## Preprocessing Data

### Dropping Columns

```python
#dropping columns that will not be good features
df = df.drop(['RowNumber','Surname','CustomerId'], axis=1)
```

### Missing Data

```python
for x in df.columns:
    if df[x].isna().sum()>0: 
        df['missing'] = np.where(df[x].isna(),1,0)
        print(x)
        display(df.corr()['missing'])
        df.drop('missing',axis=1,inplace=True)
```

Based on the above, there do not seem to be significant correlations between the missing data and the other variables, so the data is most likely missing completely at random.

#### Filling In Missing Data

```python
#filling in missing values with the median
df['Tenure'] = df['Tenure'].fillna(df['Tenure'].median())
#changing the datatype to integers
df['Tenure'] = df['Tenure'].astype('int64')
df.info()
```

##### Conclusions

Since I do not want to drop 10% of the data, I chose to fill in the missing data with the median. I choose the median because the mean and median were nearly identical, and filling in with the median allows for the conversion of `Tenure` to an integer. 

### Converting Categorical Variables To Dummy Variables

```python
#turning all categorical variables into dummy variables, while dropping one of the values 
df_dummies = pd.get_dummies(df, drop_first=True)
#looking at the new columns for the data with the dummy variables
df_dummies.info()
```

### Splitting the Data Into Train, Validate, and Test Sets

#### For Dummy Variable DF

```python
#creating target and features dataframes 
target = df_dummies['Exited']
features = df_dummies.drop('Exited',axis=1)

#splitting the training dataset from the valid and testing dataset
features_train, features_validtest, target_train, target_validtest = train_test_split(
    features, target, test_size=0.60, random_state=12345
)

#splitting the valid and testing datasets
features_valid, features_test, target_valid, target_test = train_test_split(
    features_validtest, target_validtest, test_size=0.50, random_state=12345
)
```

#### Conclusions

I split the dataframe using the commonly used ratio of 60/20/20. I did not split with stratified sampling, but that is something that could be considered when splitting the data.

### Feature Scaling

```python
#creating a list of numeric features
numeric = ['CreditScore', 'Age', 'Tenure', 'Balance','EstimatedSalary']

#scaling numeric features for the dummy dataset
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

display(features_train.head())
```

#### Conclusions

As the final part of preprocessing, I scaled all the numeric features. This is to put all the numeric features on the same scale so the models put equal weight on each of them. 

## Model Training Without Class Balance Adjustments

### Decision Tree Classification

```python
def dt_objective(trial):
    #defining ranges for decison tree optimization
    criterion = trial.suggest_categorical('criterion',['gini','entropy'])
    max_depth = trial.suggest_int('max_depth', 1, 50,1)
    max_features = trial.suggest_categorical('max_features', ['auto', 'sqrt','log2'])
    min_samples_split = trial.suggest_uniform('min_samples_split', 0, 1)
    #creating model
    model = DecisionTreeClassifier(criterion = criterion,
                                 max_depth = max_depth, max_features = max_features,
                                 min_samples_split = min_samples_split)
    model.fit(features_train,target_train)
    target_pred = model.predict(features_valid)
    #getting f1
    f1 = f1_score(target_valid, target_pred)
    return f1

start_time = time.time()
study_dt = optuna.create_study(direction='maximize',sampler=TPESampler(seed=10))
study_dt.optimize(dt_objective, n_trials=1000)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
print('Highest F1 Score',study_dt.best_value)

plot_optimization_history(study_dt)
```

#### Conclusions

For this section, I looked at a Decision Tree Classification Model. I used Optuna to find the optimal hyperparameters based on F1 score. The 4 hyperparameters I looked at were:

- **criterion**: The function to measure the quality of a split.
- **max_depth**: The maximum depth of the tree.
- **max_features**: The number of features to consider when looking for the best split
- **min_samples_split**: The minimum number of samples required to split an internal node

<cite>[Source][1]</cite> 

[1]: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

I then optimized the model based on F1 score and, after 1000 trials, the highest F1 score of a Decision Tree Classification Model was **59%**.

### SMOTE Resampling

#### What is SMOTE Resampling

"SMOTE stands for Synthetic Minority Over-Sampling Technique. SMOTE is performing the same basic task as basic resampling (creating new data points for the minority class) but instead of simply duplicating observations, it creates new observations along the lines of a randomly chosen point and its nearest neighbors. Basically you’re simulating some additional variation in the data (within the established bounds of your minority class), reducing the danger of overfitting (although not eliminating it)." <cite>[towards data science][3]</cite> 

[3]: https://towardsdatascience.com/upsampling-with-smote-for-classification-projects-e91d7c44e4bf

```python
#creating a SMOTE model
sm = SMOTE(random_state=21)
#resampling the features and targets from the training set for both the label encoded and dummy variable datasets
features_train_upsamp, target_train_upsamp = sm.fit_resample(features_train, target_train)
```

### Tomek Links

#### What are Tomek Links?

Tomek Links method uses the rule to selects the pair of observation (say, a and b) that are fulfilled these properties:

- The observation a’s nearest neighbor is b.
- The observation b’s nearest neighbor is a.
- Observation a and b belong to a different class. That is, a and b belong to the minority and majority class (or vice versa), respectively. <cite>[towards data science][4]</cite> 

[4]: https://towardsdatascience.com/imbalanced-classification-in-python-smote-tomek-links-method-6e48dfe69bbc




```python
#create a tomek link model
tl = TomekLinks()
#downsample both the label encoded and dummy training datasets using Tomek Links
features_train_downsamp, target_train_downsamp = sm.fit_resample(features_train, target_train)
```
